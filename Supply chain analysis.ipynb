# Comprehensive Analysis of Supply Chain Data

```python
#import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the data
data= pd.read_csv("supply_chain_data.csv")
```

```python
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
# No missing values 
```

```python
data['Product type'].value_counts()
```

```python
plt.figure(figsize=(4,4))
data['Product type'].value_counts().plot(kind='pie')
```

```python
data[['Product type','Revenue generated']].groupby('Product type').agg('sum').sort_values(by='Revenue generated',ascending=False)
# high revenue is generated from skincare products
```

```python
data[['Product type','Revenue generated']].groupby('Product type').agg('sum').sort_values(by='Revenue generated',ascending=False).plot(kind='bar')
```

# Revenue Analysis by region

```python
data[['Location','Revenue generated']].groupby(['Location']).agg('sum').sort_values(by='Revenue generated',ascending=False)
# high revenue is generated from Mumbai
```

Mumbai emerges as the top revenue-generating region, followed by Kolkata, while Delhi lags behind despite lower prices and shipment costs.

```python
data[['Location','Revenue generated']].groupby(['Location']).agg('sum').sort_values(by='Revenue generated',ascending=False).plot(kind='bar')
```

```python
product_revenue_by_location=data[['Location','Product type','Revenue generated']].groupby(['Location','Product type']).agg('sum')
```

```python
sns.barplot(x='Location',y='Revenue generated',hue='Product type',data=product_revenue_by_location)
```

```python
data[['Product type','Number of products sold']].groupby('Product type').agg('sum').sort_values(by='Number of products sold',ascending=False) 
# With 20731 number of products sold, the skincare product category exhibits strong sales performance
```

```python
data[['Product type','Defect rates']].groupby('Product type').agg('mean').sort_values(by='Defect rates',ascending=False)
#defect rates are high for haircare and skincare products
```

```python
data[['Transportation modes','Defect rates']].groupby('Transportation modes').agg('mean').sort_values(by='Defect rates',ascending=False)
# defect rates vs Transportation modes
```

```python
# Inspection results vs Product type
data.groupby('Product type')['Inspection results'].value_counts(normalize=True)
# For Haircare products, only few are passed in Inspection results 
```

Defect rates of 2.483150 and 2.334681, respectively, are noted in the haircare and skincare product categories, where inspection failures and high defect rates are also noted

# Supplier Evaluation

```python
# average lead time vs supplier
data[['Supplier name','Lead time']].groupby('Supplier name').agg('mean').sort_values(by='Lead time')
# supplier 1 best, supplier 3 needs improvement
```

```python
# average Defect rates vs supplier
data[['Supplier name','Defect rates']].groupby('Supplier name').agg('mean').sort_values(by='Defect rates')
# supplier 1 best, supplier 5 worst
```

```python
# supplier vs Inspection results
data.groupby('Supplier name')['Inspection results'].value_counts(normalize=True)
# supplier 4 is worst with fail rate as 66.7%
```

```python
# supplier vs Manufacturing costs
data[['Supplier name','Manufacturing costs']].groupby('Supplier name').agg('mean').sort_values(by='Manufacturing costs')
# Manufacturing costs are high for supplier 4
```

```python
data[['Supplier name','Location']].groupby(['Supplier name','Location']).agg('count')
# All suppliers are
```

1.Based on lead time metrics, defect rates, and inspection findings, Supplier 1 exhibits superior performance.

2.Supplier 4 has a notable inspection failure rate of 66.7% and high manufacturing expenses

3.Lead times and failure rates need to be improved for Supplier 3

# Shipping Logistics

```python
average_shipping_time = data['Shipping times'].mean()
print("Average Shipping Time:", average_shipping_time)
```

```python
# Average Shipping Time by Region
data[['Location','Shipping times']].groupby('Location').agg('mean').sort_values(by='Shipping times').plot(kind='bar')
```

```python
# average shipping costs
average_shipping_cost = data['Shipping costs'].mean()
print("Average Shipping Cost:", average_shipping_cost)
```

```python
data[['Location','Shipping costs']].groupby('Location').agg('mean').sort_values(by='Shipping costs').plot(kind='bar')
```

```python
data[['Location','Routes','Shipping times']].groupby(['Location','Routes']).agg('mean')
```

```python
# most frequently used shipping carriers
data['Shipping carriers'].value_counts()
```

```python
data['Shipping carriers'].value_counts().plot(kind='bar')
```

```python
# shipping times and costs across different carriers
data.groupby('Shipping carriers').agg({'Shipping times': 'mean', 'Shipping costs': 'mean'})
```

Based on shipping costs and schedules, Carrier B is the recommended shipping choice.
